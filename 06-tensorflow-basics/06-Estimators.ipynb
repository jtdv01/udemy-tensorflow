```python
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
n = 1000000
x_data = np.linspace(0.0, 10.0, n)

noise = np.random.randn(len(x_data))
y_true = (0.5* x_data) + 5 + noise

df = pd.DataFrame({'x':x_data, 'y': y_true})
```

```python
df.sample(250).plot(x='x', y='y', kind='scatter')
plt.show()
df.head()
```

```python
feat_cols = [tf.feature_column.numeric_column('x', shape=[1])]
```

```python
estimator = tf.estimator.LinearRegressor(feature_columns=feat_cols)
```

```python
x_train, x_eval, y_train, y_eval = train_test_split(x_data,y_true, test_size=0.3, random_state=123)
```

```python
print(x_train.shape)
```

```python
print(x_eval.shape)
```

```python
input_func = tf.estimator.inputs.numpy_input_fn({'x': x_train},
    y_train,
    batch_size=8, num_epochs=None, shuffle=True)
```

```python
# Don't shuffle because i want the index
train_input_func = tf.estimator.inputs.numpy_input_fn({'x': x_train}, y_train,
                                                     batch_size=8
                                                     , num_epochs=1000
                                                     , shuffle=False)
```

```python
# Don't shuffle because i want the index
eval_input_func = tf.estimator.inputs.numpy_input_fn({'x':x_eval}, y_eval,
                                                     batch_size=8
                                                     , num_epochs=1000
                                                     , shuffle=False)
```

```python
estimator.train(input_fn=input_func, steps=1000)
```

```python
train_metrics = estimator.evaluate(input_fn=train_input_func, steps=1000)
```

```python
print(train_metrics)
```

```python
eval_metrics = estimator.evaluate(input_fn=eval_input_func, steps=1000)
```

```python
print(eval_metrics)
```

# Now for predicting some new data

```python
brand_new_data = np.linspace(0,10,10)
```

```python
input_fn_predict = tf.estimator.inputs.numpy_input_fn({'x':brand_new_data}, shuffle=False)
```

```python
predictions = list(estimator.predict(input_fn=input_fn_predict))
```

```python
predictions
```

```python
pred_arr = [i['predictions'][0] for i in predictions]
```

```python
df.sample(250).plot(kind='scatter', x='x', y='y')
plt.plot(brand_new_data,pred_arr,'r*')
plt.plot(brand_new_data,pred_arr,'r-')
plt.title("Brand new predicted data")
plt.show()
```
